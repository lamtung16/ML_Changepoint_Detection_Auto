```python
import lzma
import pandas as pd
from opart_utils import opart, plot_chpnt
```

```python
with lzma.open("data/cancer/profiles.csv.xz", 'rt') as file:
    signal_df = pd.read_csv(file)
labels_df = pd.read_csv("data/cancer/labels.csv")
```

```python
seqID = '20139.16.F1'
lda = 1
chpnt = opart(signal_df[signal_df['sequenceID']==seqID]['signal'], lda)
plot_chpnt(signal_df, labels_df, seqID, chpnt, title=f"OPART with $\lambda$={lda}", figsize=(20,1))
```
