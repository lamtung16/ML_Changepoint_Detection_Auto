```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 100)
```

```python
lambda_error_df = pd.read_csv("data/cancer/evaluation.csv")
labels_df = pd.read_csv("data/cancer/labels.csv")
```

```python
labels_df['pos_label'] = (labels_df['changes'] == 1).astype(int)
labels_df['neg_label'] = (labels_df['changes'] == 0).astype(int)
labels_df = labels_df.groupby('sequenceID')[['pos_label', 'neg_label']].sum().reset_index()
```

```python
# Get the index of the minimum 'errors' value for each 'sequenceID' group
min_error_indices = lambda_error_df.groupby('sequenceID')['errors'].idxmin()

# Use these indices to select the rows from the original DataFrame
lambda_error_df_min = lambda_error_df.loc[min_error_indices].reset_index(drop=True)

# get non zeros
lambda_error_df_min_non_zero = lambda_error_df_min[lambda_error_df_min['errors'] > 0]

# Filter rows with non-zero 'errors'
lambda_error_df_min_non_zero = lambda_error_df_min[lambda_error_df_min['errors'] > 0].copy()

# Replace 'min.log.lambda' and 'max.log.lambda' with their exponential values
lambda_error_df_min_non_zero['min.lambda'] = np.exp(lambda_error_df_min_non_zero['min.log.lambda'])
lambda_error_df_min_non_zero['max.lambda'] = np.exp(lambda_error_df_min_non_zero['max.log.lambda'])

# Optionally, drop the old 'min.log.lambda' and 'max.log.lambda' columns if no longer needed
lambda_error_df_min_non_zero = lambda_error_df_min_non_zero.drop(columns=['min.log.lambda', 'max.log.lambda'])

# Sort by 'sequenceID' column
lambda_error_df_min_non_zero = lambda_error_df_min_non_zero.sort_values(by='sequenceID').reset_index(drop=True)
```

```python
final_df = pd.merge(lambda_error_df_min_non_zero, labels_df, on='sequenceID', how='left')
final_df
```
